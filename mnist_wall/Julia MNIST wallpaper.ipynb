# Wallpaper Generator

```julia
using Colors, ImageMagick, FileIO
const res = (2160, 3840)                     # UHD screen resolution
```

```julia
data = readdlm("MNIST.csv",',')[:,1:784]     # coulmn 785 is label
mnist = data[shuffle(1:size(data)[1]),:]     # shuffle data for each run
ver = res[1]รท28
hor = res[2]รท28
nd = hor * ver
digit = Array(Gray24, 28, 28)
wall = Array(Gray24, ver*28, hor*28)
h = 0
v = 0
```

```julia
for r in 1:nd-1, c in 1:784                          
    digit[cรท29+1,(c-1)%28+1] = mnist[r,c]/256
    if (c == 784) 
        wall[(v*28)+1:(v*28)+28,(h*28)+1:(h*28)+28] = digit 
        h+=1
        if (h == hor) h = 0; v += 1 end
    end
end
```
