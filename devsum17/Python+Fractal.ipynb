```python
from PIL import Image
import time
```

```python
detail = 150
h = 500
w = 500
bitmap = Image.new("RGB", (w,h))
c = -0.3819660112501051 + 0.6180339887498949j
```

```python
def fractal(z, c, maximum):
    for i in range(1, maximum):
        if abs(z) > 2:
            return i
        z = (z ** 2) + c
    return maximum;
```

```python
t0 = time.time()
for x in range(w):
    for y in range(h):
        z = complex(3*(x+1-w/2)/w, 3*(y+1-h/2)/h)
        val = fractal(z, c, detail)
        out = round(val/detail*255)
        bitmap.putpixel((x,y), (out,out,out))
t1 = time.time()

print(t1-t0)
```
