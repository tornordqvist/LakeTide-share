```r
detail <- 150
h <- 500
w <- 500
bitmap <- matrix(0, h, w)
c <- complex(real=-0.3819660112501051,imaginary=0.6180339887498949)
```

```r
fractal <- function(z, c, max) {
    for (i in 1:max) {
        if (Mod(z) > 2) return(i)
        z <- z^2 + c
    }
    return(max)
}
```

```r
t0 <- Sys.time()
for (x in 1:w) {
    for (y in 1:h) {
        z <- complex(real=(3*(x-w/2)/w), imaginary=(3*(y-h/2)/h))
        out <- fractal(z, c, detail)
        bitmap[x,y] <- out/detail
    }
}
t1 <- Sys.time()
print(t1-t0)
```

```r
x <- seq(-1, 1, length.out=w)
y <- seq(-1, 1, length.out=h)
image(x, y, bitmap)
```

```r
xmin = -1
xmax = 1
nx = 3000
ymin = -1
ymax = 1
ny = 3000
n = 150

x <- seq(xmin, xmax, length.out=nx)
y <- seq(ymin, ymax, length.out=ny)
z <- matrix(0.0, nrow=length(x), ncol=length(y))
k <- matrix(0.0, nrow=length(x), ncol=length(y))

t0 <- Sys.time()
c <- outer(x,y*1i,FUN="+")
for (rep in 1:n) { 
    index <- which(Mod(z) < 2)
    z[index] <- z[index]^2 + c[index]
    k[index] <- k[index] + 1
}
t1 <- Sys.time()
print(t1-t0)
```
